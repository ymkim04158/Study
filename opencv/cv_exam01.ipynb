```python
import cv2
cap = cv2.VideoCapture("c:\\opencv_test\\01_People.avi")
while(True):    
    bw, img = cap.read()
    #print(bw)
    orb = cv2.ORB_create()
    kp = orb.detect(img,None)
    kp, des = orb.compute(img, kp)
    img2 = cv2.drawKeypoints(img,kp,None,color=(0,0,255), flags=0)
    cv2.imshow("ORB-SLAM", img2)
    if cv2.waitKey(1)==13:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#3) 마우스 이벤트 속성값을 활용해보자. setMouseCallback(windowName, onMouse [, param]) -> None
#full(shape, fill_value, dtype+None, order ='C')

#1)ndarray를 이용한 레이아웃을 실행
# 윈도우 사이즈 변경 resizeWindow(타이틀, w, h)
import numpy as np
import cv2

def onMouse(event,x,y,flags,param):
    if event ==cv2.EVENT_LBUTTONDOWN:
        print("왼쪽 마우스 클릭",event,x,y,flags,param)
    
    elif event ==cv2.EVENT_RBUTTONDOWN:
        print("오른쪽 마우스 클릭",event,x,y,flags,param)
        
    elif event ==cv2.EVENT_LBUTTONUP:
        print("왼쪽 마우스 UP",event,x,y,flags,param)
        
    elif event ==7:
        print("왼쪽 마우스 double_Click",event,x,y,flags,param)
    
image = np.full((200,300),255,np.uint8) #배열로 이미지 생성

title01, title02 = "test01", "test02"  #윈도우타이틀 값
cv2.imshow(title01, image)
cv2.imshow(title02, image)

# 마우스 이벤트 연동
cv2.setMouseCallback("test01",onMouse) #마우스 콜백 함수
cv2.waitKey(0)  #키 이벤트
cv2.destroyAllWindows() # 객체소멸
```

```python
#4) createTrackbar 생성
#createTrackbar(const String & trackbarname, const String & winname, int * value, int count, TrackbarCallback 	onChange = 0, void * userdata = 0 )
# (트랙바 이름, 윈도우 이름, 트랙바값, 개수, onChange, 파람)

import numpy as np
import cv2

def onChange(value): #트랙바의 콜백함수
    global image # 전역변수
    
    res = value - int(image[0][0])  #트랙바 값 - 이미지 화소
    image = image + res
    print(image[0][0],res,image)
    cv2.imshow(title01,image)

image = np.zeros((200,600),np.uint8) #배열로 이미지 생성

title01 = "test01"  #윈도우타이틀 값
cv2.imshow(title01, image)

# 트랙바 연동
cv2.createTrackbar("Mytrack","test01",image[0][0],255,onChange) #마우스 콜백 함수
cv2.waitKey(0)  #키 이벤트
cv2.destroyAllWindows() # 객체소멸
```

```python
#4) createTrackbar 생성
#createTrackbar(const String & trackbarname, const String & winname, int * value, int count, TrackbarCallback 	onChange = 0, void * userdata = 0 )
# (트랙바 이름, 윈도우 이름, 트랙바값, 개수, onChange, 파람)

import numpy as np
import cv2

def onChange(value): #트랙바의 콜백함수
    global image # 전역변수
    
    res = value - int(image[0][0])  #트랙바 값 - 이미지 화소
    image = image + res
    cv2.imshow(title01,image)

def onMouse(event,x,y,flags,param): #마우스 콜백함수
    global image # 전역변수
    
    if event ==cv2.EVENT_LBUTTONDOWN:
        if(image[0][0] >= 10):
            image = image - 10
        cv2.setTrackbarPos("MyTrack","test01",image[0][0])
        print("누름")
    
    elif event ==cv2.EVENT_RBUTTONDOWN:
        print("오른쪽 마우스 클릭",event,x,y,flags,param)

image = np.zeros((300,500),np.uint8) #배열로 이미지 생성

title01 = "test01"  #윈도우타이틀 값
cv2.imshow(title01, image)

# 트랙바 연동
cv2.createTrackbar("Mytrack","test01",image[0][0],255,onChange) #마우스 콜백 함수
cv2.setMouseCallback("test01",onMouse) #마우스 콜백 함수
cv2.waitKey(0)  #키 이벤트
cv2.destroyAllWindows() # 객체소멸
```
